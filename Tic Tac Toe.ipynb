```python
from IPython.display import clear_output
import numpy as np
import random
from time import sleep

class tictactoe:
    
    def __init__(self):
        
        self.rows = int(input('For the square board give no of rows: '))
        self.mat = [['N' for j in range(self.rows)] for i in range(self.rows)]

        self.player1 = input('Choose Your Sign: X or O ?: ')

        if self.player1 == 'X':
            self.player2 = 'O'
        elif self.player1 == 'O':
            self.player2 = 'X'
        else:
            self.player1 = 'X'
            self.player2 = 'O'
            

        print(f'Player 1 - {self.player1}, Player 2 - {self.player2}')
    
    def pos_valid(self, mat, pos1, pos2):
        return not (mat[pos1][pos2] == 'X' or mat[pos1][pos2] == 'O')

    def make_board(self, mat, rows):
        x = 44
        for i in range(rows):
            for j in range(rows):
                if mat[i][j] != 'N':
                    print('\033[31m'+'| '+'\033[31m'+f'{mat[i][j]}  '+'\033[31m', end='')
                else:
                    print('\033[96m'+'|'+'\033[96m'+f'{i} {j} '+'\033[96m', end='')
            print()

    def check_winner(self, mat):
        
        global winner
        
        for i in mat:
            if len(set(i))==1:
                winner = i[0]

        for i in np.transpose(mat):
            if len(set(i))==1:
                winner = i[0]

        if len(set(np.fliplr(mat).diagonal())) == 1:
                winner = np.fliplr(mat).diagonal()[0]

        if len(set(np.array(mat).diagonal())) == 1:
                winner = np.array(mat).diagonal()[0]
        if winner == 'N':
            winner = 'No Winner !!'
        return winner
```

# Two Player - If you and your Friend want to play together

```python
prev = 'X'
winner = 'No Winner !!'

t = tictactoe()

p1, p2, mat, rows = t.player1, t.player2, t.mat, t.rows
t.make_board(mat, rows)

counter = 1
while((winner != 'X' or winner != 'O') and counter<(rows*rows)+1):
    
    pos = str(input('Input Position: '))
    if len(pos)!=2:
        print('Need double digit input, try again !! ')
        continue
    pos1 = int(pos[0])
    pos2 = int(pos[1])
    
    if t.pos_valid(mat, pos1, pos2):
        counter += 1
    else:
        print('Position Already Filled, try different place on board !!')
        continue

    mat[pos1][pos2] = prev
    clear_output()
    t.make_board(mat, rows)
    
    if t.check_winner(mat) != 'No Winner !!':
        break
    
    if prev=='X':
        prev = 'O'
    else:
        prev = 'X'
    
clear_output()
if t.check_winner(mat) != 'No Winner !!':
        print("\033[5m"+f'--- {winner} Won !! ---')
else:
    print("\033[5m"+'No Winner !!')
print()
t.make_board(mat, rows)
```
