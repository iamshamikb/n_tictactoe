```python
from IPython.display import clear_output

class TicTacToe:    
    
    def __init__(self):
        self.l = [' ']*10
        
        print('Welcome to Tic Tac Toe.')
        self.player_no = input('Chose your sign between X and O: ')
        if self.player_no == 'X': 
            self.another_player = 'O' 
        else: 
            self.another_player = 'X'
        clear_output()
        print('You chose {}, the other player chose {}.'.format(self.player_no, self.another_player))        

    def show_board(self):
        
        print('-'*13)
        print('| {} | {} | {} |'.format(self.l[7], self.l[8], self.l[9]))
        print('-'*13)

        print('| {} | {} | {} |'.format(self.l[4], self.l[5], self.l[6]))
        print('-'*13)

        print('| {} | {} | {} |'.format(self.l[1], self.l[2], self.l[3]))
        print('-'*13) 
        
    def board(self, player_inp, turn):
        clear_output()
        if self.l[player_inp] == ' ':
            if turn == 'X':
                self.l[player_inp] = 'X'
            else:
                self.l[player_inp] = 'O'

            self.show_board()
            return self.check()
        else:
            print('Invalid input, try another space.')
            self.show_board()
            return True
        
    def check(self):
        if (self.l[1] == self.l[2] == self.l[3] == 'X') or (self.l[4] == self.l[5] == self.l[6] == 'X') \
        or (self.l[7] == self.l[8] == self.l[9] == 'X'):
            return False
        elif (self.l[1] == self.l[2] == self.l[3] == 'O') or (self.l[4] == self.l[5] == self.l[6] == 'O') \
        or (self.l[7] == self.l[8] == self.l[9] == 'O'):
            return False
        elif (self.l[1] == self.l[5] == self.l[9] == 'X') or (self.l[7] == self.l[5] == self.l[3] == 'X'):
            return False
        elif (self.l[1] == self.l[5] == self.l[9] == 'O') or (self.l[7] == self.l[5] == self.l[3] == 'O'):
            return False
        elif (self.l[1] == self.l[4] == self.l[7] == 'X') or (self.l[2] == self.l[5] == self.l[8] == 'X') \
        or (self.l[3] == self.l[6] == self.l[9] == 'X'):
            return False
        elif (self.l[1] == self.l[4] == self.l[7] == 'O') or (self.l[2] == self.l[5] == self.l[8] == 'O') \
        or (self.l[3] == self.l[6] == self.l[9] == 'O'):
            return False
        else:
            return True

play = True


t = TicTacToe()
turn = t.player_no

winner = False

count = 1
while(play and count<10):
    if count==1:
        t.show_board()
    player_inp = input('For player {}: '.format(turn))
    play = t.board(int(player_inp), turn)
    if not play:
        print('Congratulations! Player {} Won'.format(turn))
        winner = True
    if turn == 'X':
        turn = 'O'
    else:
        turn = 'X'
    count+=1

if not winner:
    print('Nobody wins !!')
```
