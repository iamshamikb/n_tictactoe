```python
from IPython.display import clear_output
import numpy as np

def pos_valid(mat, pos1, pos2):
    return not (mat[pos1][pos2] == 'X' or mat[pos1][pos2] == 'O')

def make_board(mat, rows):
    x = 44
    for i in range(rows):
        for j in range(rows):
            if mat[i][j] != 'N':
                print('\033[31m'+'| '+'\033[31m'+f'{mat[i][j]}  '+'\033[31m', end='')
            else:
                print('\033[96m'+'|'+'\033[96m'+f'{i} {j} '+'\033[96m', end='')
        print()
        
def start():
    rows = int(input('For the square board give no of rows: '))
    mat = [['N' for j in range(rows)] for i in range(rows)]
    
    player1 = input('Choose Your Sign: X or O ?: ')

    if player1 == 'X':
        player2 = 'O'
    else:
        player2 = 'X'
        player1 = 'O'

    print(f'Player 1 - {player1}, Player 2 - {player2}')
    return player1, player2, mat, rows

def check_winner(mat):
    global winner
    for i in mat:
        if len(set(i))==1:
            winner = i[0]

    for i in np.transpose(mat):
        if len(set(i))==1:
            winner = i[0]

    if len(set(np.fliplr(mat).diagonal())) == 1:
            winner = np.fliplr(mat).diagonal()[0]

    if len(set(np.array(mat).diagonal())) == 1:
            winner = np.array(mat).diagonal()[0]
    if winner == 'N':
        winner = 'No Winner !!'
    return winner

prev = 'X'
winner == 'No Winner !!'

p1, p2, mat, rows = start()
make_board(mat, rows)

counter = 1
while((winner != 'X' or winner != 'O') and counter<(rows*rows)+1):
    
    pos = str(input('Input Position: '))
    if len(pos)!=2:
        print('Need double digit input, try again !! ')
        continue
    pos1 = int(pos[0])
    pos2 = int(pos[1])
    
    if pos_valid(mat, pos1, pos2):
        counter += 1
    else:
        print('Position Already Filled, try different place on board !!')
        continue

    mat[pos1][pos2] = prev
    clear_output()
    make_board(mat, rows)
    
    if check_winner(mat) != 'No Winner !!':
        break
    
    if prev=='X':
        prev = 'O'
    else:
        prev = 'X'
    
clear_output()
if check_winner(mat) != 'No Winner !!':
        print("\033[5m"+f'--- {winner} Won !! ---')
else:
    print("\033[5m"+'No Winner !!')
print()
make_board(mat, rows)
```
